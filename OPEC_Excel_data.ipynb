```python
import pandas as pd
from datetime import datetime as dt
```

```python
OPEC_April_MOMRreport_file = 'data/MOMR Appendix Tables (April 2020).xlsx'

OPEC_demand_supply_balance_df = pd.read_excel(OPEC_April_MOMRreport_file,sheet_name=1)
OPEC_demand_supply_balance_column_names = OPEC_demand_supply_balance_df.iloc[3,1:15].values.copy()

OPEC_demand_supply_balance_cleaned_df = OPEC_demand_supply_balance_df.iloc[4:,1:15].copy()
OPEC_demand_supply_balance_cleaned_df.reset_index(drop=True,inplace=True)
OPEC_demand_supply_balance_cleaned_df.columns = OPEC_demand_supply_balance_column_names
OPEC_demand_supply_balance_cleaned_df
```

```python
OPEC_world_rig_counts_df = pd.read_excel(OPEC_April_MOMRreport_file,sheet_name=5)

OPEC_world_rig_counts_df_column_names_series = OPEC_world_rig_counts_df.iloc[4,1:]
OPEC_world_rig_counts_df_column_names = OPEC_world_rig_counts_df_column_names_series.values.copy()
OPEC_world_rig_counts_df_column_names[4] = 'Change 2019/18'
OPEC_world_rig_counts_df_column_names[9] = dt.strftime(OPEC_world_rig_counts_df_column_names[9],'%b %d')
OPEC_world_rig_counts_df_column_names[10] = dt.strftime(OPEC_world_rig_counts_df_column_names[10],'%b %d')
OPEC_world_rig_counts_df_column_names[11] = 'Change Mar/Feb'

OPEC_world_rig_counts_cleaned_df = OPEC_world_rig_counts_df.iloc[5:,1:].copy()
OPEC_world_rig_counts_cleaned_df.reset_index(drop=True,inplace=True)
OPEC_world_rig_counts_cleaned_df.columns = OPEC_world_rig_counts_df_column_names
OPEC_world_rig_counts_cleaned_df
```
